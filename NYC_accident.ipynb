```python
df["Year"] = df["CRASH DATE"].dt.year
df["Month"] = df["CRASH DATE"].dt.month
df["Day"] = df["CRASH DATE"].dt.day
df["Weekday"] = df["CRASH DATE"].dt.day_name()
df["Hour"] = df["CRASH TIME"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap
import datetime
```

```python
df["CRASH TIME"] = pd.to_datetime(df["CRASH TIME"], errors='coerce')
df["Hour"] = df["CRASH TIME"].dt.hour
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x="Hour", palette="coolwarm")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Hour of the Day in NYC (2020)")
plt.xticks(rotation=45)
plt.show()
```

```python
df = df.dropna(subset=["Hour"])

df.loc[:, "Hour"] = df["Hour"].astype(int)

def categorize_time(hour):
    if 0 <= hour < 6:
        return "Late Night (12 AM - 6 AM)"
    elif 6 <= hour < 12:
        return "Morning (6 AM - 12 PM)"
    elif 12 <= hour < 18:
        return "Afternoon (12 PM - 6 PM)"
    else:
        return "Evening (6 PM - 12 AM)"

df.loc[:, "Time of Day"] = df["Hour"].apply(categorize_time)

plt.figure(figsize=(12, 7))
sns.violinplot(
    x="Time of Day", y="Hour", data=df, palette="Set2", order=[
        "Late Night (12 AM - 6 AM)", "Morning (6 AM - 12 PM)",
        "Afternoon (12 PM - 6 PM)", "Evening (6 PM - 12 AM)"],
    inner="quartile", linewidth=2
)

plt.xlabel("Time of Day", fontsize=12, fontweight="bold")
plt.ylabel("Crash Time (Hour)", fontsize=12, fontweight="bold")
plt.title("Accident Density Across Different Parts of the Day", fontsize=14, fontweight="bold")

plt.xticks(rotation=20, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.6) 

plt.show()
```

```python
df = df.dropna(subset=["LATITUDE", "LONGITUDE"])
```

```python
from folium.plugins import HeatMap, MarkerCluster
nyc_map = folium.Map(location=[40.7128, -74.0060], zoom_start=12, tiles="cartodbpositron")

heat_data = df[["LATITUDE", "LONGITUDE"]].values.tolist()

HeatMap(heat_data, radius=15, blur=10, max_zoom=12).add_to(nyc_map)

marker_cluster = MarkerCluster().add_to(nyc_map)
for lat, lon in zip(df["LATITUDE"], df["LONGITUDE"]):
    folium.CircleMarker(
        location=[lat, lon],
        radius=3,
        color="red",
        fill=True,
        fill_color="red",
        fill_opacity=0.6
    ).add_to(marker_cluster)

nyc_map.save("NYC_Detailed_Accident_Heatmap.html")

print("Heatmap saved as 'NYC_Detailed_Accident_Heatmap.html'. Open it in a browser to view.")
```

```python
accident_by_borough = df["BOROUGH"].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=accident_by_borough.index, y=accident_by_borough.values, palette="viridis")
plt.xlabel("Borough")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Borough in NYC (2020)")
plt.xticks(rotation=45)
ax = sns.barplot(x=accident_by_borough.index, y=accident_by_borough.values, palette="viridis")
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')
plt.show()
```

```python
injuries_vs_fatalities = df[["NUMBER OF PERSONS INJURED", "NUMBER OF PERSONS KILLED"]].sum()

plt.figure(figsize=(8, 5))
ax = injuries_vs_fatalities.plot(kind="bar", color=["blue", "red"])
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Total Injuries vs. Fatalities in NYC Accidents (2020)")
plt.xticks(rotation=45)

# Add labels on bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

```python
accident_severity = df[["NUMBER OF PEDESTRIANS INJURED", "NUMBER OF PEDESTRIANS KILLED",
                        "NUMBER OF CYCLIST INJURED", "NUMBER OF CYCLIST KILLED",
                        "NUMBER OF MOTORIST INJURED", "NUMBER OF MOTORIST KILLED"]].sum()

plt.figure(figsize=(10, 6))
ax = accident_severity.plot(kind="bar", color=["blue", "red", "green", "purple", "orange", "brown"])
plt.xlabel("Road User Type")
plt.ylabel("Count")
plt.title("Accident Severity by Road User Type in NYC (2020)")
plt.xticks(rotation=45)

# Add labels on bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```
